## LowerCasing

```python
# Define the target columns
cols_to_lower =['left_stance',
 'left_stance_specific',
 'left_newsTitle',
 'left_newBody',
 'left_opinionTag',
 'center_stance',
 'center_stance_specific',
 'center_newsTitle',
 'center_newBody',
 'center_opinionTag',
 'right_stance',
 'right_stance_specific',
 'right_newsTitle',
 'right_newBody',
 'right_opinionTag',
 'issue',
 'topic',
 'roundup']

# Apply lowercasing using the .str accessor
for col in cols_to_lower:
    # We cast to string first to handle any mixed types (like News_ID)
    # then apply lower()
    df[col] = df[col].astype(str).str.lower()

# Quick check to verify
print(df[cols_to_lower].head())
```

```python
import re

def clean_for_transformer(text):
    if not isinstance(text, str):
        return ""

    # 1. Remove digital noise but keep sentence markers
    text = re.sub(r'https?://\S+|www\.\S+', '', text) # Remove URLs
    text = re.sub(r'@\S+', '', text)                  # Remove Mentions

    # 2. Keep only alphanumeric and basic punctuation [.,!?;:]
    # This allows the model to see the "structure" of the news article
    text = re.sub(r'[^a-zA-Z0-9\s.,!?;:]', '', text)

    # 3. Standardize whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
# 1. Define the columns to be cleaned for the Transformer
target_cols = ['left_stance',
 'left_stance_specific',
 'left_newsTitle',
 'left_newBody',
 'left_opinionTag',
 'center_stance',
 'center_stance_specific',
 'center_newsTitle',
 'center_newBody',
 'center_opinionTag',
 'right_stance',
 'right_stance_specific',
 'right_newsTitle',
 'right_newBody',
 'right_opinionTag',
 'issue',
 'topic',
 'roundup']

# 2. Apply the 'clean_for_transformer' function
for col in target_cols:
    print(f"Cleaning {col} for Transformer...")
    df[col] = df[col].apply(clean_for_transformer)

# 3. Quick Sanity Check
# Let's see how the first row of News_Body looks now
print("\nSample Cleaned Text:")
```

```python
import pkg_resources
from symspellpy import SymSpell, Verbosity

# 1. Initialize SymSpell with max edit distance of 2
sym_spell = SymSpell(max_dictionary_edit_distance=2, prefix_length=7)

# 2. Load the included English frequency dictionary
dictionary_path = pkg_resources.resource_filename(
    "symspellpy", "frequency_dictionary_en_82_765.txt"
)
# term_index=0 (word), count_index=1 (frequency)
sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)

# 3. Define the fast correction function
def symspell_correct(text):
    if not isinstance(text, str) or not text.strip():
        return text

    # lookup_compound handles multi-word sentences and preserves spaces
    suggestions = sym_spell.lookup_compound(text, max_edit_distance=2)
    return suggestions[0].term

# 4. Apply to your bias project columns
cols_to_fix = ['left_stance',
 'left_stance_specific',
 'left_newsTitle',
 'left_newBody',
 'center_stance',
 'center_stance_specific',
 'center_newsTitle',
 'center_newBody',
 'right_stance',
 'right_stance_specific',
 'right_newsTitle',
 'right_newBody',
 'issue',
 'topic',
 'roundup']
for col in cols_to_fix:
    print(f"Fast-correcting {col}...")
    df[col] = df[col].apply(symspell_correct)
```

```python
import pandas as pd

# 1. Define the stance-to-label mapping
label_map = {
    'left': 'liberal',
    'right': 'conservative',
    'center': 'center'
}

# 2. Create the three buckets with specific stance and mapped labels
# We use the 'stance_specific' columns for the Stance column as you requested
left_df = df[['left_newsTitle', 'left_newBody', 'left_stance_specific', 'issue', 'topic', 'roundup']].copy()
left_df.columns = ['Title', 'News_Body', 'Stance', 'issue', 'topic', 'roundup']
left_df['Label'] = label_map['left']

right_df = df[['right_newsTitle', 'right_newBody', 'right_stance_specific', 'issue', 'topic', 'roundup']].copy()
right_df.columns = ['Title', 'News_Body', 'Stance', 'issue', 'topic', 'roundup']
right_df['Label'] = label_map['right']

center_df = df[['center_newsTitle', 'center_newBody', 'center_stance_specific', 'issue', 'topic', 'roundup']].copy()
center_df.columns = ['Title', 'News_Body', 'Stance', 'issue', 'topic', 'roundup']
center_df['Label'] = label_map['center']

# 3. Combine them and create a unique News_ID for each original story
df_final = pd.concat([left_df, right_df, center_df], ignore_index=True)

# Generate News_ID based on the original row groups
# (This ensures all 3 rows from the same original row share one ID)
df_final['News_ID'] = df_final.index % len(df)

# 4. Sort Serially: News_ID 0 (L,R,C), News_ID 1 (L,R,C)...
# We create a custom sort order for the Label
custom_order = {'liberal': 0, 'conservative': 1, 'center': 2}
df_final['order'] = df_final['Label'].map(custom_order)

df_final = df_final.sort_values(by=['News_ID', 'order']).reset_index(drop=True)

# 5. Set final column order
final_cols = ['News_ID', 'Title', 'News_Body', 'Stance', 'Label', 'issue', 'topic', 'roundup']
df_final = df_final[final_cols]

# print("Transformation Complete. Preview of serial grouping:")
# print(df_final.head(6))
```
